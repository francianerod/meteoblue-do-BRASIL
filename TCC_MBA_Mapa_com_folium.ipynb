# **1) IMPORTANDO PACOTES E ORGANIZANDO DADOS**

```python
# Importando pacotes
import pandas as pd
import folium
```

```python
# Fazendo um dicionário com lat e lon
latitude = {'Met': -23.211283, 
            'Bin': -23.209431, 
            'Aer': -23.224739,
            'Peq': -23.201461,
            'IAG': -23.559251,
            'IEAV':-23.251728,
            'Novadutra': -23.353560
            }

longitude = {'Met': -45.860278, 
            'Bin': -45.880862, 
            'Aer': -45.862521,
            'Peq': -45.873773,
            'IAG': -46.734083,
            'IEAV':-45.857773,
            'Novadutra': -46.181000
            }
```

```python
# Transformando os dicionários em dataframes
lat_data = pd.DataFrame(list(latitude.items()), columns=['Sensor','Latitude'])
lon_data = pd.DataFrame(list(longitude.items()), columns=['Sensor','Longitude'])

# União dos dois dataframe 
dataset = lat_data.merge(lon_data, on = "Sensor", how = "left")

# Mostrando o resultado do dataset
display(dataset)
```

# **2) CONSTRUINDO MAPA COM FOLIUM**

```python
# Construindo mapa com folium

map = folium.Map(location=[dataset.Latitude.mean(), 
                           dataset.Longitude.mean()], 
                 zoom_start=10, 
                 control_scale=True
                 )
```

```python
#Adicionando pontos do Mapa

for index, location_info in dataset.iterrows():
    folium.Marker([location_info["Latitude"], 
                   location_info["Longitude"]],
                   icon=folium.Icon(icon='Sensor'),
                   popup=location_info["Sensor"]).add_to(map)
map
```
