### **OBSERVAÇÕES TÉCNICAS**

1. Geada meteorológica: Na meteorologia, define-se ocorrência de geada quando há deposição de gelo sobre plantas e objetos expostos ao relento. No entanto, mesmo com a formação de gelo sobre as plantas pode não haver morte dos tecidos vegetais, por elas estarem em repouso vegetativo.

2. Geada agrícola: Na Agronomia, entende-se geada como fenômeno atmosférico que provoca a morte das plantas ou de suas partes (folhas, caule, frutos, ramos), em função da baixa temperatura do ar, que acarreta congelamento dos tecidos vegetais, havendo ou não, formação de gelo sobre a planta.

3. Para as culturas milho, sorgo e algodão temperatura crítica será utilizada no intervalo de -1°C e -2°C, independente do estadio fenológico.

4. Pode ser usado tanto para previsão na segmentação diária como para registro de estação meteorológica (física ou virtual)

### **BASE DE DADOS**

```python
# Passo 1: Abrir as bibliotecas
import pandas as pd
```

```python
# Passo 2: Entrada de dados
# Entrada de dados: Temperatura mínima local (5h)
# Dados da Basiléia em janeiro de 2022.

temp_min = [2.3, 2.3, 11.1, 5.4, 0.8, -2.1, -1.1, 2.6, 1.5 ,-0.5, ]
df = pd.DataFrame(temp_min, columns=["temp_min"])
display(df)
```

### **GEADA METEOROLÓGICA**

```python
# Função da geada meteorológica
def func1(column):

    if column['temp_min'] > 4.0:
       return "Sem geada meteorológica"

    elif column['temp_min'] > 3.0 and column['temp_min'] <= 4.0:
      return "Geada Meteorológica Fraca"

    elif column['temp_min'] > 1.0 and column['temp_min'] <= 3.0:
      return "Geada Meteorológica Moderada"

    elif column['temp_min'] <= 1.0:
       return "Geada Meteorológica Forte"

    else: 
       return

#Aplicando a função em uma nova coluna
df['geada_met'] = df.apply(func1, axis=1)
display(df)
```

### **GEADA AGRICOLA** 

```python
# Equação da temperatura de relva
df['temp_relva'] = df['temp_min'] - 4
#df
```

```python
# Função da geada agrícola para Milho, Sorgo e Algodão
def func2(column):

    if column['temp_relva'] > -1.0:
       return "Sem geada agrícola"

    elif column['temp_relva'] > -2.0 and column['temp_relva'] <= - 1.0:
      return "Possibilidade de geada agrícola"

    elif column['temp_relva'] < -2.0:
       return "Geada agrícola"

    else: 
       return

#Aplicando a função em uma nova coluna
df['geada_agri'] = df.apply(func2, axis=1)
display(df)
```
